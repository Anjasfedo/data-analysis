# Load Libraries

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# Basic Graph

## Line Chart

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]

# Basic plotting
plt.plot(x, y)
plt.show()
```

```python
plt.plot(x, y)

# Title & Label
plt.title("My First Graph",
          fontdict={
              "fontsize": 20,
              "fontweight": "bold",
              "color": "red"
          })
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")

plt.show()
```

```python
plt.plot(x, y)

# Tick
plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8])

plt.show()
```

```python
plt.plot(x,
         y,
         label="Line 1",
        #  color="red",
         color="#0000FF",

         linewidth=5,
         linestyle="--",

         marker="o",
         markersize=10,
         markerfacecolor="yellow",
         markeredgecolor="green",
         markeredgewidth=2
         )

# Legend
plt.legend()

plt.show()
```

```python
# Shorthand notation
# fmt = '[color][marker][line]'

plt.plot(x, y, "ro--", label="Line 1")

plt.legend()

plt.show()
```

### 2 Lines

```python
x2 = np.arange(0,4.5,0.5)
y2 = x2**2

plt.plot(x, y, "r", label="First Line")
plt.plot(x2, y2, "y", label="Second Line")

plt.legend()

plt.show()
```

```python
plt.plot(x, y, "r", label="First Line")
plt.plot(x2[4:], y2[4:], "y--", label="Second Line")

plt.legend()

plt.show()
```

### Resize graph

```python
plt.figure(figsize=(5,3)) # x,y

plt.plot(x, y, "r", label="First Line")
plt.plot(x2, y2, "y--", label="Second Line")

plt.legend()

plt.show()
```

```python
plt.figure(figsize=(5,4))

plt.plot(x, y, "r", label="First Line")
plt.plot(x2, y2, "y--", label="Second Line")

plt.legend()

plt.show()
```

```python
plt.figure(figsize=(6,3))

plt.plot(x, y, "r", label="First Line")
plt.plot(x2, y2, "y--", label="Second Line")

plt.legend()

plt.show()
```

### Save Graph

```python
plt.figure(figsize=(6,3))

plt.plot(x, y, "r", label="First Line")
plt.plot(x2, y2, "y--", label="Second Line")

plt.legend()

plt.savefig("mygraph.png")

plt.show()
```

## Bar Chart

```python
labels = ["A", "B", "C"]
values = [1,4,2]

plt.bar(labels, values)

plt.show()
```

```python
plt.bar(labels, values)

plt.figure(figsize=(6,4))

plt.show()
```

```python
bars = plt.bar(labels, values)

bars[0].set_hatch("/")
bars[1].set_hatch("O")
bars[2].set_hatch("*")

plt.show()
```

```python
bars = plt.bar(labels, values)

patterns = ["/", "O", "*"]

for bar in bars:
    bar.set_hatch(patterns.pop(0))

plt.show()
```

## Line Chart

```python
gas = pd.read_csv("gas_prices.csv")

plt.figure(figsize=(8,5))

plt.plot(gas.Year, gas.USA, "b.-", label="USA")
plt.plot(gas.Year, gas.Canada, "r.-", label="Canada")
plt.plot(gas.Year, gas["South Korea"], "g.-", label="South Korea")
plt.plot(gas.Year, gas.Australia, "y.-", label="Australia")
plt.plot(gas.Year, gas.France, "m.-", label="France")
plt.plot(gas.Year, gas.Italy, "c.-", label="Italy")
plt.plot(gas.Year, gas.Japan, "k.-", label="Japan")
plt.plot(gas.Year, gas.Germany, "k.-", label="Germany")
plt.plot(gas.Year, gas.Mexico, "b.-", label="Mexico")
plt.plot(gas.Year, gas.UK, "r.-", label="UK")

plt.title("Gas Prices Over Time (USD)", fontdict={"fontsize": 18})
plt.xlabel("Year")
plt.ylabel("USD")

plt.legend()

plt.xticks(gas.Year[::3].tolist())

plt.savefig("gas_price_figure.png")

plt.show()
```
