# Load Libraries

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# Basic Graph

## Line Chart

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]

# Basic plotting
plt.plot(x, y)
plt.show()
```

```python
plt.plot(x, y)

# Title & Label
plt.title("My First Graph",
          fontdict={
              "fontsize": 20,
              "fontweight": "bold",
              "color": "red"
          })
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")

plt.show()
```

```python
plt.plot(x, y)

# Tick
plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8])

plt.show()
```

```python
plt.plot(x,
         y,
         label="Line 1",
        #  color="red",
         color="#0000FF",

         linewidth=5,
         linestyle="--",

         marker="o",
         markersize=10,
         markerfacecolor="yellow",
         markeredgecolor="green",
         markeredgewidth=2
         )

# Legend
plt.legend()

plt.show()
```

```python
# Shorthand notation
# fmt = '[color][marker][line]'

plt.plot(x, y, "ro--", label="Line 1")

plt.legend()

plt.show()
```

### 2 Lines

```python
x2 = np.arange(0,4.5,0.5)
y2 = x2**2

plt.plot(x, y, "r", label="First Line")
plt.plot(x2, y2, "y", label="Second Line")

plt.legend()

plt.show()
```

```python
plt.plot(x, y, "r", label="First Line")
plt.plot(x2[4:], y2[4:], "y--", label="Second Line")

plt.legend()

plt.show()
```

### Resize graph

```python
plt.figure(figsize=(5,3)) # x,y

plt.plot(x, y, "r", label="First Line")
plt.plot(x2, y2, "y--", label="Second Line")

plt.legend()

plt.show()
```

```python
plt.figure(figsize=(5,4))

plt.plot(x, y, "r", label="First Line")
plt.plot(x2, y2, "y--", label="Second Line")

plt.legend()

plt.show()
```

```python
plt.figure(figsize=(6,3))

plt.plot(x, y, "r", label="First Line")
plt.plot(x2, y2, "y--", label="Second Line")

plt.legend()

plt.show()
```

### Save Graph

```python
plt.figure(figsize=(6,3))

plt.plot(x, y, "r", label="First Line")
plt.plot(x2, y2, "y--", label="Second Line")

plt.legend()

plt.savefig("mygraph.png")

plt.show()
```
