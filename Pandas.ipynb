# Loading in DataFrame from Files

## Download the file

```python
import requests

def download_file(url, filename):
    response = requests.get(url)
    if response.status_code == 200:
        with open(filename, 'wb') as file:
            file.write(response.content)
```

```python
download_file(url="https://raw.githubusercontent.com/KeithGalli/complete-pandas-tutorial/master/warmup-data/coffee.csv",
              filename="coffee.csv")
```

```python
# .apply()
bios["height_category"] = bios["height_cm"].apply(lambda x: "Tall" if x > 200 else "Short")
```

```python
bios[bios["height_category"] == "Tall"]
```

```python
def categorize_athlete(row):
    if row["height_cm"] < 175 and row["weight_kg"] < 70:
        return "Ligthweight"
    elif row["height_cm"] < 185 or row["weight_kg"] <= 80:
        return "Mediumweight"
    else:
        return "Heavyweight"

bios["category"] = bios.apply(categorize_athlete, axis=1)
```
